# Notebook to the Birth rate analysis
## Data by Centers for Disease Control (CDC)

```python
# imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

```python
# trattament of missing values
births = pd.read_csv('births.csv')
print(births.head())
#births['day'].fillna(0, inplace=True) - Depreccated
replace_value = 0
births['day'] = [replace_value if np.isnan(x) else x for x in births['day']]
births['day'] = births['day'].astype(int)
print(births.head())
```

# Calculate decade

```python
births['decade'] = 10 * (births['year'] // 10)
births.pivot_table('births', index='decade', columns='gender', aggfunc='sum')
print(births.head())
```

## Create a plot date

```python
birth_decade = births.pivot_table('births', index='decade', columns='gender', aggfunc='sum')
birth_decade.plot()
plt.ylabel('Total births per year')
plt.show()
```
